```python
#Problem-3: Array operations:

#Sol 1-
import numpy as np

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])

add_result = x + y
print("Addition of x and y:\n", add_result)


#Sol 2-
subtract_result = x - y
print("Subtraction of x and y:\n", subtract_result)


#Sol 3-
multiply_x = x * 2
multiply_y = y * 3
print("x multiplied by 2:\n", multiply_x)
print("y multiplied by 3:\n", multiply_y)


#Sol 4-
square_x = np.square(x)
square_y = np.square(y)
print("Square of each element of x:\n", square_x)
print("Square of each element of y:\n", square_y)


#Sol 5-
v = np.array([9, 10])
w = np.array([11, 12])

dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)

print("Dot product between v and w:", dot_vw)
print("Dot product between x and v:\n", dot_xv)
print("Dot product between x and y:\n", dot_xy)
```
